```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
url = 'https://bit.ly/2cLzoxH'
brecha = pd.read_csv(url)
brecha
```

```python
brecha.info()
```

```python
brecha.lifeExp.hist(bins=50)
```

```python
#Se pueden cambiar los parametros 
brecha.lifeExp.hist(bins=100, grid=False, xlabelsize=12, ylabelsize=12)
plt.title( 'Histograma de pandas',fontsize=15)
plt.xlabel('Esperanza de vida',fontsize=14)
plt.ylabel('Frecuencia',fontsize=14)
plt.show()
```

```python
brecha.gdpPercap.hist(bins=100, grid=False)
plt.title( 'Histograma de pandas',fontsize=15)
plt.xlabel('Ingreso per capita',fontsize=14)
plt.ylabel('Frecuencia',fontsize=14)
plt.show()
```

```python
brecha.gdpPercap.hist(bins=1000, grid=False)
plt.title( 'Histograma de pandas',fontsize=15)
plt.xlabel('Ingreso per capita',fontsize=14)
plt.ylabel('Frecuencia',fontsize=14)
plt.xscale('log')
plt.show()
```

```python
import seaborn as sns
```

```python
#Grafica de distribuciÃ³n acumulativa 
sns.distplot(brecha.lifeExp)
```

```python
sns.distplot(brecha.lifeExp, color='green', kde=True, bins=100)
plt.title( 'Histograma de seaborn',fontsize=15)
plt.xlabel('Esperanza de vida',fontsize=14)
plt.ylabel('Frecuencia',fontsize=14)
plt.show()
```

```python
#Comparando la brecha de esperanza de vida enatre dos paÃ­ses
df = brecha[brecha.continent == 'Africa']
sns.distplot(df.lifeExp, label='Africa')
df = brecha[brecha.continent== 'Americas']
sns.distplot(df.lifeExp, label='Americas')
plt.title( 'Histograma de seaborn',fontsize=15)
plt.xlabel('Esperanza de vida',fontsize=14)
plt.ylabel('Frecuencia',fontsize=14)
plt.legend()
plt.show()
```

```python
df = brecha[brecha.continent == 'Africa']
sns.distplot(df.lifeExp, label='Africa',hist=False)
df = brecha[brecha.continent== 'Americas']
sns.distplot(df.lifeExp, label='Americas', hist=False)
plt.title( 'Histograma de seaborn',fontsize=15)
plt.xlabel('Esperanza de vida',fontsize=14)
plt.ylabel('Frecuencia',fontsize=14)
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-white')
data = np.random.randn(1000)
plt.hist(data)
```

```python
#Creamos tres conjuntos con distribuciÃ³n normal y los comparamos 
x = np.random.normal(0,0.7,1000)
y = np.random.normal(2,1.5,1000)
z = np.random.normal(-3,1,1000)

dic = dict(histtype='stepfilled',alpha=0.5,bins=50)
plt.hist(x,**dic)
plt.hist(y,**dic)
plt.hist(z,**dic)

plt.show()
```

```python
#Histograma de dos dimensiones
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean, cov,10000).T


plt.hist2d(x,y,bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('total por contenedor')
```

```python
plt.hexbin(x,y,gridsize=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('total por contenedor')
```
